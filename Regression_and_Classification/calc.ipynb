```python
def sigmoid(z):
    """
    Compute the sigmoid of z

    Args:
        z (ndarray): A scalar, numpy array of any size.

    Returns:
        g (ndarray): sigmoid(z), with the same shape as z
         
    """

    g = 1/(1+np.exp(-z))
   
    return g
```

```python
y_t = np.array([sigmoid(0),sigmoid(8)-1,sigmoid(5),sigmoid(0)-1])

x_T = np.array([[1,-1,-2],[1,2,6],[1,3,3],[1,2,-2]])
1/4 * (x_T.T@y_t)
```
